```python
import numpy as np
import matplotlib.pyplot as plt
filename = 'data.txt'

data = np.loadtxt(filename)

print(data)
force = data[:,3]
dist = data[:,2]

#plt.title('Fjarlægð sem fall af krafti')
plt.plot(dist,force)
plt.xlabel('Fjarlægð frá yfirborði [$\mu m$]')
plt.ylabel('Kraftur [nN]')
plt.grid()
plt.savefig('forceDist.png')
```
